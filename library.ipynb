```python
import sqlite3
import pandas as pd

def initialize_db():
    conn = sqlite3.connect('Library.db')
    cursor = conn.cursor()

    sql = """
        CREATE TABLE IF NOT EXISTS BookTable (
            bookID INTEGER PRIMARY KEY,
            Title TEXT,
            Author TEXT,
            Genre TEXT,
            Year INTEGER, 
            Series INTEGER,
            Height REAL,
            Width REAL
        )
    """
    
    cursor.execute(sql)
    conn.commit()
    conn.close()
    print('Library table has been created.')

#ADD COLUMN TO DB FUNCTION?

def add_book(book_details):
    # NEED TO ADD A CHECK TO SEE IF BOOK ALREADY IN LIBRARY.
    conn = sqlite3.connect('Library.db')
    cursor = conn.cursor()

    insert_query = "INSERT INTO BookTable (Title, Author, Genre, Year, Series, Height, Width) VALUES (?, ?, ?, ?, ?, ?, ?)"
    data = book_details
    
    cursor.execute(insert_query, data)
    conn.commit()
    conn.close()

    print(f'{book_details[0]} by {book_details[1]} was succesfully added to the library.')

def add_several_books(books_list):
    for book_details in books_list:
        add_book(book_details)
```
